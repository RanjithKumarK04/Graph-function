```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import math
from sklearn.preprocessing import Imputer

#Define a graph function to upload required dataset and directory to save the results
def graph(data,columns=0,directory= os.getcwd()):
        #Check if user provided number of columns, if not pick all the columns from the dataset
    if columns==0:
        columns = data.columns.values
        data = data.loc[:,columns]
    else:
        data = data.loc[:,columns]
        #Create a new variable 'categorical_cols' to store the categorical variables from the dataset
    categorical_cols = data.select_dtypes(include=['object']).copy().columns.values
    
        #Create a new variable 'numerical_cols' to store the numerical variables from the dataset
    numerical_cols = np.delete(columns,np.searchsorted(columns,categorical_cols))
    sns.set(style="darkgrid",palette="muted", color_codes=True)
    c=0
    length=data.shape[0]
        # Calculting the length of data.shape[0] for the given dataset
    while length !=0:
        c +=1
        length = int(length/10)
    for cols in numerical_cols:
        
        #Getting all the categorical variables header from dataset
        if data[cols].value_counts().count() < max(math.floor(data.shape[0]/(10**(c-1))),15):
            categorical_cols = np.append(categorical_cols,cols)
            
        #Dropping all the categorical variables header names for getting numerical variables header names
    numerical_cols = data.drop(categorical_cols,axis=1).columns.values
    print("Numerical_cols:",numerical_cols)
    print("Categorical_cols",categorical_cols)
    
        #Asking the user if user want to do imputation on variables of dataset
    opt1= input("Do You Want to impute data?(y/n)")
    if opt1.lower() == "y":
        if np.sum(data.loc[:,numerical_cols].isnull().sum()) > 0:
        
        #If User says 'Yes' performing 'missing value imputation' on the numerical variables
            mean_imputer = Imputer(missing_values='NaN', strategy='mean', axis=0)
            mean_imputer = mean_imputer.fit(data.loc[:,numerical_cols])
            data.loc[:,numerical_cols] = mean_imputer.transform(data.loc[:,numerical_cols].values)
        else:
            print("Sorry there are no NaN or Null values in numerical_cols can't do imputation.. continuing to plotting graphs")
        if np.sum(data.loc[:,categorical_cols].isnull().sum())>0:
            
        #If User says 'Yes' performing 'missing value imputation' on the categorical variables
            mean_imputer = Imputer(missing_values='NaN', strategy='most_frequent', axis=0)
            mean_imputer = mean_imputer.fit(data.loc[:,categorical_cols])
            data.loc[:,categorical_cols] = mean_imputer.transform(data.loc[:,categorical_cols].values)
        else:
             print("Sorry there are no NaN or Null values in categorical_cols can't do imputation.. continuing to plotting graphs")
    else:
        
        # If user says 'No' then drop all the num values from the dataset
        data=data.dropna()
        
        #Askign user for creating a smooth curve for the plots
    opt2 = input("Do you want smoothed curve?(y/n):")
    
        #Creating histogram and box plot for all numerical variables in the dataset
    for cols in numerical_cols:
        counts, bins = np.histogram(data[cols])
        plt.figure(figsize=[10,7])
        grid = plt.GridSpec(2, 2, wspace=0.4, hspace=1.5)
        plt.subplot(grid[0,0])
        plt.suptitle("Histogram and Box plots for "+cols,size=20,weight=5,y=1.2)
        if opt2.lower()=="y":
            
        #if user say "yes", function will create smoothed curves using seaborn library
            sns.distplot(data[cols],bins,color='orange')
        else:
            
        #If user says 'No' then boxplots will be created for all the numerical variables in the dataset
            sns.distplot(data[cols],bins,kde=False)
        plt.subplot(grid[0,1])
        sns.boxplot(x=data[cols],hue=data[cols], orient= 'v',color='orange')
        plt.subplot(grid[1,:2],frame_on=False,in_layout=False)
        plt.axis('off')
        
        #Creating summary for each variable below the graph in the .jpeg file
        plt.figtext(0.5,0.4,'Summary for {} column'.format(cols),ha='center',va='center',size=18)
        colus = ('count','mean','std','min','25%','50%','75%','max')
        cell=list(data[cols].describe().values)
        cell_=[str(round(i,2)) for i in cell]
        cell_text=[]
        cell_text.append(cell_)
        ytable = plt.table(cellText=cell_text,colLabels=colus,loc='center',cellLoc = 'center', rowLoc = 'center')
        ytable.set_fontsize(15)
        ytable.scale(1, 4)
        
        #Saving every plot in the location provided by the user
        plt.savefig(directory+'/'+cols+ '_histBar.png', bbox_inches='tight')
        
    sns.despine(left=True)
    
        #Creating barplot for all categorical variables in the dataset
    for cols in categorical_cols:
        plt.figure(figsize=[10,7])
        grid = plt.GridSpec(2, 2, wspace=0.4, hspace=1.5)
        plt.subplot(grid[0,:2])
        plt.suptitle("Bar Plot for "+cols,size=20,weight=5,y=1.2)
        sns.barplot(data[cols].value_counts().index,data[cols].value_counts().values,alpha=0.9, hue = data[cols].value_counts().index)
        plt.ylabel('Number of Occurrences', fontsize=12)
        plt.subplot(grid[1,:2],frame_on=False,in_layout=False)
        plt.axis('off')
        
        #Creating summary for each variable below the graph in the .jpeg file
        plt.figtext(0.5,0.4,'Summary for {} column'.format(cols),ha='center',va='center',size=18)
        colus = ('count','min','max')
        count = str(data[cols].count())
        maxi = str(data[cols].mode())
        mii = data[cols].value_counts()
        mii = mii.sort_index(ascending = True)
        mini = str(mii.index[0])
        cell_text=[[count,mini,maxi]]
        ytable = plt.table(cellText=cell_text,colLabels=colus,loc='center',cellLoc = 'center', rowLoc = 'center')
        ytable.set_fontsize(15)
        ytable.scale(1, 4)
        
        #Saving every plot in the location provided by the user
        plt.savefig(directory+'/'+cols+'.png', bbox_inches='tight')
```
